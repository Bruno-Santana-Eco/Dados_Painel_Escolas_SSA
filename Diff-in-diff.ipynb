### Comparando os colegios.

```python
from fuzzywuzzy import fuzz

# Garantir que os valores so string e no nulos
df_uniao['NO_ENTIDADE'] = df_uniao['NO_ENTIDADE'].fillna('').astype(str)
df_uniao['name_school'] = df_uniao['name_school'].fillna('').astype(str)

# Aplica a funo de similaridade
df_uniao['similaridade'] = df_uniao.apply(
    lambda row: fuzz.ratio(row['NO_ENTIDADE'], row['name_school']),
    axis=1
)

# Filtra os casos com baixa similaridade (por exemplo, abaixo de 85%)
df_diferentes = df_uniao[df_uniao['similaridade'] < 75]
```

```python
variaveis_controle = [
    'IN_REFEITORIO', 'IN_LABORATORIO_INFORMATICA', 'IN_LABORATORIO_CIENCIAS', 'IN_BIBLIOTECA',
    'IN_AUDITORIO', 'IN_QUADRA_ESPORTES_COBERTA', 'IN_SALA_MULTIUSO', 'IN_EQUIP_LOUSA_DIGITAL',
    'IN_COMPUTADOR', 'IN_INTERNET']
formula = 'SOMA_MAT ~ pos + tratamento + did + ' + ' + '.join(variaveis_controle)

modelo = smf.ols(formula, data=df).fit()
print(modelo.summary())
```

```python
variaveis_controle = [
    'IN_REFEITORIO', 'IN_LABORATORIO_INFORMATICA', 'IN_LABORATORIO_CIENCIAS', 'IN_BIBLIOTECA',
    'IN_AUDITORIO', 'IN_QUADRA_ESPORTES_COBERTA', 'IN_SALA_MULTIUSO', 'IN_EQUIP_LOUSA_DIGITAL',
    'IN_COMPUTADOR', 'IN_INTERNET']

formula = (
    'SOMA_MAT ~ pos + tratamento + did + ' +
    ' + '.join(variaveis_controle) +
    ' + C(TP_DEPENDENCIA) + C(TP_ATIVIDADE_COMPLEMENTAR)' 
)

modelo = smf.ols(formula, data=df).fit()
print(modelo.summary())
```
